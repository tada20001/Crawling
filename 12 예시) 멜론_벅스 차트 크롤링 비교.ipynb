## 멜론

```python
melon_chart_url = "https://www.melon.com/chart/"

headers = {'User-Agent': user_agent}
res = requests.get(melon_chart_url, headers=headers)
res  # 거부. 헤더 지정 필요
```

```python
html = res.text
soup = BeautifulSoup(html, "html.parser")
```

```python
import re
#song_list = soup.select("[id^=lst]")
song_list = soup.select('.lst50, .lst100') #,로 두부분 선택 가능
len(song_list)
```

```python
tag = song_list[1]
song_id = tag["data-song-no"]
title = tag.select_one('[href*=playSong]').text
artist_tag = tag.select_one('[href*=goArtistDetail]')
artist = artist_tag.text
album = tag.select_one('.wrap_song_info [href*=goAlbum]').text
song_id, title, artist, album
```

```python
# 정규표현식을 이용하여 아티스트 ID 추출
import re
matched = re.search(r'(\d+)', artist_tag['href'])
artist_id = matched.group(1)
```

```python
# function
def get_id_string(tag):
    matched = re.search(r'(\d+)', tag['href'])
    return matched.group(1)
```

```python
songs_list = []

for tag in song_list:
    song_id = tag["data-song-no"]
    title = tag.select_one('[href*=playSong]').text

    artist_tag = tag.select_one('[href*=goArtistDetail]')
    artist = artist_tag.text
    artist_id = get_id_string(artist_tag)

    album_tag = tag.select_one('.wrap_song_info [href*=goAlbum]')
    album = album_tag.text
    album_id =get_id_string(album_tag)

    songs_list.append({
        'song_id': song_id,
        'song_title': title,
        'artist_id' : artist_id,
        'artist_name': artist,
        'album_id': album_id,
        'album_name': album,
    })
```

## 벅스 뮤직

```python
bugs_chart_url = "https://music.bugs.co.kr/chart"

res = requests.get(bugs_chart_url)
res
```

```python
html = res.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
tag_list = soup.select('.byChart tr')
tag_list.remove(tag_list[0])  # 첫번째 태그 제외
# 또는
tag_list = soup.select('.byChart tr[trackid]')  # 이렇게 지정하면 tr 중 trackid 속성이 있는 것만 추출
len(tag_list)
```

```python
album_id = tag_list[0]['albumid']
artist_id = tag_list[0]['artistid']
movie_id = tag_list[0]['mvid']
track_id = tag_list[0]['trackid']
title = tag_list[0].select_one('.title a').text
artist_name = tag_list[0].select_one('.left .artist a').text
album_name = tag_list[0].select_one(".left .album").text
title, artist_name, album_name
```

```python
# 또는 노래 속성 따로 만들기....
tag = tag_list[0]

song_attrs = []

for tag in tag_list:
    attrs = {key: value for key, value in tag.attrs.items() if key.endswith('id')}  
    # tag의 속성 중 key 값이 id로 끝나는 경우만 추출
    attrs.update(dict(
        song_title = tag.select_one('.title').text.strip(),
        artist_name = tag.select_one('.artist').text.strip(),
        album_name = tag.select_one('.album').text.strip(),
    ))
    song_attrs.append(attrs)
```
