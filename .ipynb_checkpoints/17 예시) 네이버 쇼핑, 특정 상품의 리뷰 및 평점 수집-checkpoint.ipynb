## 함수 만들기

```python
def get_review_list(mv_mid, page=1, reviewSort='accuracy', reviewType="all", 
                    topicCode=None, score=None):
    review_list_url = "https://search.shopping.naver.com/detail/review_list.nhn"
    params = {
    'nvMid': mv_mid,
    'page': page,
    'reviewSort': reviewSort,  # 랭킹순(accuracy), 최신순(registeration)
    'reviewType': reviewType,       # 전체(all), 포토리뷰(photo)
    'topicCode': topicCode,      # 주제 전체인 경우 필요없음
    'score': score,                 # 평점도 1~5 선택 가능, 지정안하면 전체 추출
    'ligh': 'true',
}
    if topicCode:
        params["topicCode"] = topicCode
    if score:
        params["score"] = score

    res = requests.post(review_list_url, params=params)
    html = res.text
    soup = BeautifulSoup(html, "html.parser")
    tag_list = soup.select(".thumb_nail")

    
    review_list = []

    for tag in tag_list:
        info_tag = tag.select('.info_cell')
        review_list.append(dict(
            star_point = tag.select_one('.curr_avg').text,
            shopping_mall = info_tag[0].text,
            author = info_tag[1].text,
            register_date = info_tag[2].text,
            subject = tag.select_one('.subjcet, .subject').text.strip() ,
            content = tag.select_one('.atc').text.strip(),
        ))
    return review_list
```

```python
import time
review_list = []

for page in range(1, 4):
    review_list.extend(get_review_list(df.nvmid[0], page=page))
    time.sleep(0.1)

len(review_list)
```

```python
df_review_list = pd.DataFrame(review_list)
print(df_review_list.shape)
df_review_list.head()
```
