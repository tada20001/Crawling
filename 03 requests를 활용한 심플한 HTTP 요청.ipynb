## 요청시에 QueryString 인자 지정

```python
url = "https://www.ntis.go.kr/sims/pjtinfo/pjtMainInfo.do?pjtInfoVo.pjtId=1711019127"
response = requests.get(url)
```

다음과 같이 params 인자로 구분하면 좀더 편리하다. (dict/list/tuple 모두 지원한다.)

```python
params = {"pjtInfoVo.pjtId": "1711019127"}
```

key 중복을 허용하기 위한 방법: 원래 HTTP에서는 Parameter 이름 중복을 지원한다.

```python
params = [("pjtInfoVo.pjtId", "1711019127")] # 중복지원 가능

response = requests.get("https://www.ntis.go.kr/sims/pjtinfo/pjtMainInfo.do", params=params)
print(response.text)
```

### 이 문제를 해결하기 위해서 다음과 같이 헤더 설정을 해 줄 수 있다.

```python
# http://getright.com/ 에서 확인 가능
request_headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36',
    'Referer': 'http://www.melon.com'
}
url = "https://www.melon.com/chart/index.htm"
res = requests.get(url, headers=request_headers)
```

requests 라이브러리에서 기본 User-Agent 값은 'python-requests/버전'이다. 서버에 따라 User-Agent 값으로 요청 거부여부를 결정하기도 한다.

## 손쉬운 HTTP 요청 (POST)
주로 데이터 추가/수정/삭제 목적으로 사용한다. 즉, 데이터 변화를 요구하는 요청이다.

### POST 요청의 활용
* 게시판에 글쓰기/수정/삭제
    * 웹서비스 성격에 따라 그에 맞춰 방법을 적용해야 한다.
    * request 요청을 통해 파일 업로드도 가능

#### 웹 API 로의 요청

### data 인자
post 요청에서 사용

#### 일반적인 Form 요청
dict/tuple/list 타입으로 data 지정

```python
data = {"k1": 'v1', "k2": "v2"}
response = requests.post('http://httpbin.org/post', data=data)
```

### JSON API
JSON 문자열로 직렬화하여 요청이 필요할 때

```python
# 방법 1. json 포맷 문자열로 변환한 후, data 인자로 지정
json_string = json.dumps(data, ensure_ascii=False) # import json 필요
response = requests.post("http://httpbin.org/post", data=json_string)
```

```python
# 방법 2. 객체를 json 인자로 지정하면 내부적으로 json.dumps 처리
response = requests.post("http://httpbin.org/post",json=data)
```

```python
response.text
```

### 파일 업로드 요청
multipart/form-data 인코딩

```python
files = {
    'photo1': open('f1.jpg', 'rb'), # 데이터만 전송
    'photo2': open('f2.jpg', 'rb'),
    'photo3': ('f3.jpg', open('f3.jpg', 'rb'), 'image/jpeg', {'Expires': '0'}),  # 파일명, 데이터, 파일형식
}
data = {'k1': 'v1'}
response = requests.post("http://httpbin.org/post", files=files, data=data)
```
